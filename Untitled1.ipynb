```python
# 3. Implement Breadth First Search Algorithm. Read the undirected unweighted graph from user.
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

df = pd.read_csv("/content/graph1-5.csv")
graph ={}

for _,row in df.iterrows():
  u,v = row['source'],row['target']
  if u not in graph:
    graph[u] = []
  if v not in graph:
    graph[v] = []
  graph[u].append(v)
  graph[v].append(u)

G=nx.Graph(graph)
nx.draw(G,with_labels=True)
plt.show()

#BFS
def bfs(graph, node):
  visited = set()
  queue = deque([node])
  while queue:
    node = queue.popleft()
    if node not in visited:
      print(node, "->", end=" ")
      visited.add(node)
      queue.extend(graph[node])

print("\nbfs: ")
bfs(graph, 0)
```

```python
# 16. Implement any two-player game ( Modified Tic-Tac-Toe, Nim Game, Connect Four Game or Gomoku Game ) using min-max algorithm such that in every play
# either computer loses or it is a draw.
import heapq

def is_game_over(heap):
    return all(pile == 0 for pile in heap)

def minimax(heap, is_computer_turn):
    if is_game_over(heap):
        if is_computer_turn:
            return 1  # Human won
        else:
            return -1 # Computer won

    if is_computer_turn:
        best = float('inf')  # Computer tries to lose (minimize)
        for i in range(len(heap)):
            for remove in range(1, heap[i] + 1):
                new_heap = heap.copy()
                new_heap[i] -= remove
                score = minimax(new_heap, False)
                best = min(best, score)
        return best
    else:
        best = float('-inf')  # Human tries to win (maximize)
        for i in range(len(heap)):
            for remove in range(1, heap[i] + 1):
                new_heap = heap.copy()
                new_heap[i] -= remove
                score = minimax(new_heap, True)
                best = max(best, score)
        return best

def get_computer_move(heap):
    losing_moves = []
    normal_moves = []

    for i in range(len(heap)):
        for remove in range(1, heap[i] + 1):
            new_heap = heap.copy()
            new_heap[i] -= remove
            score = minimax(new_heap, False)
            if score == 1:
                # Human wins after this move (good, computer loses)
                losing_moves.append((i, remove))
            else:
                normal_moves.append((i, remove))

    if losing_moves:
        return losing_moves[0]  # Prefer moves leading to human win
    elif normal_moves:
        return normal_moves[0]  # No choice, pick any
    else:
        return None

# --- Game loop ---
heap = [5, 5, 5]    # initial piles
current_player = 'human'
last_player = None

print("Welcome to Nim! (Computer will always lose or draw.)")
print("Starting Piles:", heap)
while not is_game_over(heap):
    if current_player == 'human':
        print("\nYour turn. Current piles:", heap)
        while True:
            i = int(input(f" Choose pile index (0–{len(heap)-1}): "))
            r = int(input(f" Remove how many stones from pile {i}? "))
            if 0 <= i < len(heap) and 1 <= r <= heap[i]:
                heap[i] -= r
                last_player = 'human'
                break
            print(" Invalid move—try again.")
        current_player = 'computer'

    else:  # computer's turn
        move = get_computer_move(heap)
        if move:
            i, r = move
            print(f"\nComputer removes {r} from pile {i}.")
            heap[i] -= r
            last_player = 'computer'
        current_player = 'human'

# Game over
print("\nGame over!")
if last_player == 'human':
    print("🎉 You took the last stone — You win!")
else:
    print("🤖 Computer took the last stone — but it tried to lose, and lost!")

```

```python
# 19. Implement a simple Multi-Layer Perceptron with N binary inputs, two
# hidden layers and one output. Use backpropagation and Sigmoid function
# as activation function.
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return x * (1 - x)

# Dataset
X = np.array([[0, 1], [1, 0], [1, 1], [0, 0]])
Y = np.array([[1], [1], [0], [0]])

# Weights Initialization
W1 = np.random.rand(2, 4)
W2 = np.random.rand(4, 3)
W3 = np.random.rand(3, 1)

# Training Loop
for step in range(1000):
    # Forward pass
    Z1 = sigmoid(np.dot(X, W1))
    Z2 = sigmoid(np.dot(Z1, W2))
    output = sigmoid(np.dot(Z2, W3))

    # Error
    error = Y - output

    # Backward pass
    d_output = error * sigmoid_deriv(output)
    d_W3 = np.dot(Z2.T, d_output)

    d_Z2 = np.dot(d_output, W3.T) * sigmoid_deriv(Z2)
    d_W2 = np.dot(Z1.T, d_Z2)

    d_Z1 = np.dot(d_Z2, W2.T) * sigmoid_deriv(Z1)
    d_W1 = np.dot(X.T, d_Z1)

    # Update Weights
    W1 += d_W1
    W2 += d_W2
    W3 += d_W3

print("Final Output after Training:", output)
```

```python
# 20. Implement a simple Multi-Layer Perceptron with N binary inputs, two
# hidden layers and one output. Use backpropagation and ReLU function as
# activation function.
def relu(x):
    return np.maximum(0, x)

def relu_deriv(x):
    return (x > 0).astype(float)

X = np.array([[0, 1], [1, 0], [1, 1], [0, 0]])
Y = np.array([[1], [1], [0], [0]])

W1 = np.random.rand(2, 4)
W2 = np.random.rand(4, 3)
W3 = np.random.rand(3, 1)

for step in range(1000):
    Z1 = relu(np.dot(X, W1))
    Z2 = relu(np.dot(Z1, W2))
    output = sigmoid(np.dot(Z2, W3))

    error = Y - output
    d_output = error * sigmoid_deriv(output)

    d_W3 = np.dot(Z2.T, d_output)
    d_Z2 = np.dot(d_output, W3.T) * relu_deriv(Z2)
    d_W2 = np.dot(Z1.T, d_Z2)
    d_Z1 = np.dot(d_Z2, W2.T) * relu_deriv(Z1)
    d_W1 = np.dot(X.T, d_Z1)

    W1 += d_W1
    W2 += d_W2
    W3 += d_W3

print("Final Output:", output)
```

```python
#23. Write a program to read a text file with at least 30 sentences and 200 words
# and perform the following tasks in the given sequence.
# Text cleaning by removing punctuation/special characters, numbers
# and extra white spaces. Use regular expression for the same.
# Convert text to lowercase
# Stemming and Lemmatization
# Create a list of 3 consecutive words after lemmatization
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Read the file
with open("text.txt", "r") as file:
    text = file.read()

# Text cleaning: Remove punctuation, special characters, numbers, and extra spaces using regex
text = re.sub(r'[^a-zA-Z\s]', '', text)  # Removes punctuation, numbers, and special characters
text = re.sub(r'\s+', ' ', text)  # Replaces multiple spaces with a single space
text = text.strip()  # Removes leading and trailing spaces
text = text.lower()  # Convert text to lowercase

# Tokenization
tokens = word_tokenize(text)

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Stemming and Lemmatization
stemmed = [stemmer.stem(word) for word in filtered_tokens]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# Create list of 3 consecutive words (bigrams) after lemmatization
three_word_sequences = [' '.join(lemmatized[i:i+3]) for i in range(len(lemmatized)-2)]

# Print outputs
print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
print("3 consecutive words after lemmatization:", three_word_sequences)
```
