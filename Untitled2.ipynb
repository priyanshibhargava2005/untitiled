```python
# 1.Implement Recursive Depth First Search Algorithm. Read the undirected unweighted graph from a .csv file.
import pandas as pd
import networkx as nx
import pandas as pd             # Pandas to read the CSV file
import networkx as nx           # NetworkX for graph operations
import matplotlib.pyplot as plt  # Matplotlib for plotting the graph
from collections import deque    # Deque for queue implementation (though not used here)

# Reading the CSV file into a DataFrame
df = pd.read_csv("/content/graph1-5.csv")

# Initializing an empty dictionary to represent the graph
graph = {}

# Iterating through each row of the DataFrame
for _, row in df.iterrows():
    u, v = row['source'], row['target']  # Extracting source and target nodes from each row
    # Adding the source node to the graph if it's not already present
    if u not in graph:
        graph[u] = []
    # Adding the target node to the graph if it's not already present
    if v not in graph:
        graph[v] = []
    # Appending the target node to the source node's adjacency list
    graph[u].append(v)
    # Appending the source node to the target node's adjacency list (since it's an undirected graph)
    graph[v].append(u)

# Creating a graph using NetworkX from the adjacency list (graph)
G = nx.Graph(graph)

# Drawing the graph with labels to show the structure of the graph
nx.draw(G, with_labels=True)
plt.show()  # Displaying the graph plot

# Defining the recursive DFS function
def dfs(graph, visited, node):
    # If the node has not been visited yet
    if node not in visited:
        visited.add(node)  # Mark the node as visited
        print(node, "->", end=" ")  # Print the node (this can be used for visualization or debugging)
        # Recursively visit each adjacent node (neighbor) of the current node
        for m in graph[node]:
            if m not in visited:  # If the neighbor hasn't been visited
                dfs(graph, visited, m)  # Recurse on that neighbor

# Initialize an empty set to keep track of visited nodes
visited = set()

# Print the starting message for DFS
print("dfs recursive: ")

# Start the DFS traversal from node 0 (or any other starting node)
dfs(graph, visited, 0)
```

```python
# 2. Implement Non-Recursive Depth First Search Algorithm. Read the undirected unweighted graph from user.
import pandas as pd             # Pandas to read the CSV file
import networkx as nx           # NetworkX for graph operations
import matplotlib.pyplot as plt  # Matplotlib for plotting the graph
from collections import deque    # Deque for stack implementation (though we use a list for stack here)

# Reading the CSV file into a DataFrame
df = pd.read_csv("/content/graph1-5.csv")

# Initializing an empty dictionary to represent the graph
graph = {}

# Iterating through each row of the DataFrame
for _, row in df.iterrows():
    u, v = row['source'], row['target']  # Extracting source and target nodes from each row
    # Adding the source node to the graph if it's not already present
    if u not in graph:
        graph[u] = []
    # Adding the target node to the graph if it's not already present
    if v not in graph:
        graph[v] = []
    # Appending the target node to the source node's adjacency list
    graph[u].append(v)
    # Appending the source node to the target node's adjacency list (since it's an undirected graph)
    graph[v].append(u)

# Creating a graph using NetworkX from the adjacency list (graph)
G = nx.Graph(graph)

# Drawing the graph with labels to show the structure of the graph
nx.draw(G, with_labels=True)
plt.show()  # Displaying the graph plot

# Defining the non-recursive DFS function using an explicit stack
def dfs_non(graph, node):
    visited = set()  # Initialize a set to keep track of visited nodes
    stack = [node]   # Initialize the stack with the starting node

    # Continue as long as there are nodes in the stack
    while stack:
        node = stack.pop()  # Pop the top node from the stack

        # If the node has not been visited yet
        if node not in visited:
            print(node, "->", end=" ")  # Print the node (for visualization or debugging)
            visited.add(node)  # Mark the node as visited

            # Add all unvisited adjacent nodes to the stack (reversed to simulate DFS order)
            stack.extend(reversed(graph[node]))

# Print the starting message for the non-recursive DFS
print("\ndfs non recursive: ")

# Start the non-recursive DFS traversal from node 0 (or any other starting node)
dfs_non(graph, 0)
```

```python
# 3. Implement Breadth First Search Algorithm. Read the undirected unweighted graph from user.
# Importing necessary libraries
import pandas as pd             # Pandas for reading the CSV file
import networkx as nx           # NetworkX for graph-related operations
import matplotlib.pyplot as plt  # Matplotlib for plotting the graph
from collections import deque    # Deque to implement the queue for BFS

# Reading the CSV file into a DataFrame (graph edges)
df = pd.read_csv("/content/graph1-5.csv")

# Initializing an empty dictionary to represent the graph
graph = {}

# Iterating through each row of the DataFrame (representing edges between nodes)
for _, row in df.iterrows():
    u, v = row['source'], row['target']  # Extracting the source and target nodes
    # Adding the source node to the graph if it's not already present
    if u not in graph:
        graph[u] = []
    # Adding the target node to the graph if it's not already present
    if v not in graph:
        graph[v] = []
    # Adding the target node to the source node's adjacency list (for undirected graph)
    graph[u].append(v)
    # Adding the source node to the target node's adjacency list (for undirected graph)
    graph[v].append(u)

# Creating a graph using NetworkX from the adjacency list (graph dictionary)
G = nx.Graph(graph)

# Drawing and displaying the graph with labels for nodes
nx.draw(G, with_labels=True)
plt.show()  # Display the graph plot

# Defining the Breadth-First Search (BFS) function
def bfs(graph, node):
    visited = set()          # A set to track visited nodes
    queue = deque([node])    # Initialize the queue with the starting node (BFS starts from this node)

    # While there are still nodes in the queue to explore
    while queue:
        node = queue.popleft()  # Dequeue a node from the left side of the queue
        if node not in visited:  # If the node has not been visited yet
            print(node, "->", end=" ")  # Print the node (this can be used for visualization)
            visited.add(node)         # Mark the node as visited
            queue.extend(graph[node])  # Add all unvisited neighbors of the node to the queue

# Print the starting message for BFS
print("\nbfs: ")

# Start BFS traversal from node 0 (or any other starting node)
bfs(graph, 0)
```

```python
# 4. Implement Best First Search Algorithm. Read the directed unweighted graph and the heuristic values from user.
# Import necessary libraries
import pandas as pd             # Pandas for reading the CSV file
import networkx as nx           # NetworkX for graph-related operations
import matplotlib.pyplot as plt  # Matplotlib for plotting the graph
import heapq                     # Heapq for implementing the priority queue

# Load the directed graph from the CSV file
df1 = pd.read_csv('/content/graph1.csv')  # Reading the graph data from CSV file
graph1 = {}  # Initialize an empty dictionary to store the graph

# Iterating through each row of the CSV file (each row represents an edge)
for _, row in df1.iterrows():
    u, v = row['source'].strip(), row['target'].strip()  # Extracting source and target nodes (stripping any extra spaces)
    # If the source node is not in the graph, add it
    if u not in graph1:
        graph1[u] = []
    # Append the target node to the adjacency list of the source node (directed edge)
    graph1[u].append(v)

# Create a directed graph using NetworkX from the adjacency list
G1 = nx.DiGraph(graph1)

# Draw the graph with node labels
nx.draw(G1, with_labels=True)
plt.show()  # Display the graph

# Take heuristic values from the user for each node
heuristics1 = {}  # Dictionary to store heuristic values for each node
for node in graph1:
    # Prompt the user to enter the heuristic value for each node
    heuristic = int(input(f"Enter heuristic value for node {node}: "))
    heuristics1[node] = heuristic  # Store the heuristic value for the node

# Best First Search (BFS) Algorithm
def best_first_search_directed_unweighted(graph, heuristics, start, goal):
    visited = set()  # Set to track visited nodes
    # Initialize the priority queue with the start node, using its heuristic value for sorting
    queue = [(heuristics[start], start)]

    # While there are still nodes to explore in the queue
    while queue:
        _, node = heapq.heappop(queue)  # Pop the node with the smallest heuristic value (best candidate)

        # If the node has not been visited yet
        if node not in visited:
            print(node, end=" ")  # Print the node (for visualization or debugging)
            visited.add(node)  # Mark the node as visited

            # If the goal node is found, stop the search
            if node == goal:
                break

            # Explore all neighbors of the current node
            for neighbor in graph.get(node, []):
                # Get the heuristic value for the neighbor, defaulting to infinity if not found
                neighbor_heuristic = heuristics.get(neighbor, float('inf'))  # Default to infinity if no heuristic is provided
                # Push the neighbor into the queue with its heuristic value
                heapq.heappush(queue, (neighbor_heuristic, neighbor))

# Take the start and goal node as input from the user
start = input("\nEnter start node: ")  # Starting node for the search
goal = input("Enter goal node: ")    # Goal node for the search

# Output the result of the Best First Search algorithm
print("\nBest First Search - Directed Unweighted:")
best_first_search_directed_unweighted(graph1, heuristics1, start, goal)  # Call the Best First Search function
```

```python
# 8. Implement A* algorithm. Read directed weighted graph and heuristic values from a .csv file.
# Import necessary libraries
import heapq               # heapq is used to create a priority queue (min-heap)
import csv                 # csv module is used to read and parse CSV files

# --- Common A* Function ---
def a_star(graph, heuristics, start, goal):
    # Initialize open set with the starting node.
    # The open set contains tuples of the form (f = g + h, g, node)
    open_set = [(heuristics[start], 0, start)]  # f = g + h (f: total cost, g: cost to reach node, h: heuristic)

    # A set to track visited nodes to avoid reprocessing
    visited = set()

    # Start the search
    while open_set:
        # Pop the node with the lowest f value (i.e., the one with the lowest estimated cost)
        f, g, node = heapq.heappop(open_set)

        # If the goal is reached, print the result and return
        if node == goal:
            print(f"\nReached {node} with total cost {g}")  # g is the total cost to reach this node
            return

        # If the node has not been visited yet
        if node not in visited:
            visited.add(node)  # Mark the node as visited
            print(node, end=" -> ")  # Print the node being visited

            # Explore neighbors of the current node
            for neighbor, weight in graph.get(node, []):  # Get the neighbors of the node
                # Calculate the new f value for the neighbor: f = g + weight + heuristic
                heapq.heappush(open_set, (g + weight + heuristics.get(neighbor, 0), g + weight, neighbor))

# --- 1. Read Directed Graph from CSV ---
def read_directed_graph_from_csv(file):
    graph = {}  # Dictionary to store the graph (adjacency list)
    heuristics = {}  # Dictionary to store the heuristic values for each node

    with open(file, 'r') as f:  # Open the CSV file
        reader = csv.reader(f)  # Create a CSV reader object
        next(reader)  # Skip the header row of the CSV

        # Read each row of the CSV file
        for row in reader:
            src, dest, weight, heuristic = row  # Extract source, destination, edge weight, and heuristic
            weight = int(weight)  # Convert weight to an integer
            heuristic = int(heuristic)  # Convert heuristic value to an integer

            # Add the destination node to the adjacency list of the source node
            if dest != '':  # Skip empty destination (in case there are no outgoing edges)
                graph.setdefault(src, []).append((dest, weight))  # Add an edge from src to dest with the given weight

            # Set the heuristic value for the source node
            heuristics[src] = heuristic
            # If the destination node doesn't have a heuristic value, assign it a default value of 0
            heuristics.setdefault(dest, 0)

    return graph, heuristics  # Return the graph and the heuristic values

# --- Running for Directed Graph (CSV) ---
print("\n\n====== A* Search: Directed Weighted Graph (from CSV) ======")

# Read the directed graph and its heuristics from the CSV file
graph_directed, heuristics_directed = read_directed_graph_from_csv('a_star_directed.csv')

# Call the A* function to find the shortest path from node 'A' to node 'D'
a_star(graph_directed, heuristics_directed, 'A', 'D')
```

```python
# 9. Implement A* algorithm. Read directed weighted graph and heuristic values from user.
import heapq  # heapq is used to create a priority queue (min-heap) for A* algorithm

# --- Common A* Function ---
def a_star(graph, heuristics, start, goal):
    # Initialize the open set with the start node,
    # (f = g + h, g, node, path) where:
    # f = total estimated cost (g + h), g = cost to reach the node, h = heuristic value, path = path taken so far
    open_set = [(heuristics[start], 0, start, [start])]  # Priority queue (min-heap)

    visited = set()  # Set to keep track of visited nodes to avoid revisiting them

    while open_set:
        # Pop the node with the lowest f value (i.e., the one with the lowest estimated cost)
        f, g, node, path = heapq.heappop(open_set)

        # If we reach the goal node, print the path and the total cost
        if node == goal:
            print(" -> ".join(path))  # Join the path into a string to display the sequence of nodes
            print(f"\nReached {node} with total cost {g}")  # Print the total cost to reach the goal
            return

        # If the node has not been visited, process it
        if node not in visited:
            visited.add(node)  # Mark this node as visited

            # Explore neighbors of the current node
            for neighbor, weight in graph.get(node, []):  # Get the neighbors and their edge weights
                # Calculate the new g value (cost to reach the neighbor)
                new_g = g + weight

                # Calculate the new f value: f = g + h (where h is the heuristic of the neighbor)
                new_f = new_g + heuristics.get(neighbor, 0)

                # Add the neighbor to the open set with the updated f, g, and path
                heapq.heappush(open_set, (new_f, new_g, neighbor, path + [neighbor]))

# --- User Input for Directed Graph ---
print("\n\n====== A* Search: Directed Weighted Graph (User Input) ======")

# Initialize dictionaries to store the graph and heuristics
graph_directed_input = {}
heuristics_directed_input = {}

# Input number of edges for the graph
n = int(input("Enter number of edges: "))

# Read edges from the user input
for _ in range(n):
    # Input source node, destination node, and edge weight
    src = input("Enter source node: ")
    dest = input("Enter destination node: ")
    weight = int(input(f"Enter weight from {src} to {dest}: "))

    # Add the edge to the graph (source -> destination with given weight)
    graph_directed_input.setdefault(src, []).append((dest, weight))

# Collect all unique nodes (including destinations) in the graph
nodes = set(graph_directed_input.keys())  # Start by adding source nodes
for edges in graph_directed_input.values():
    for dest, _ in edges:
        nodes.add(dest)  # Add destination nodes to the set

# Input heuristic values for each node
for node in nodes:
    heuristics_directed_input[node] = int(input(f"Enter heuristic value for {node}: "))

# Input the start and goal nodes
start = input("Enter start node: ")
goal = input("Enter goal node: ")

# Call the A* function to find the shortest path from start to goal
a_star(graph_directed_input, heuristics_directed_input, start, goal)
```

```python
# 20. Implement a simple Multi-Layer Perceptron with N binary inputs, two
# hidden layers and one output. Use backpropagation and ReLU function as
# activation function.
import numpy as np  # Import NumPy library for numerical operations

# ReLU activation function: Outputs the input if positive, otherwise outputs 0
def relu(x):
    return np.maximum(0, x)  # ReLU function: returns max(0, x) for each element in x

# Derivative of ReLU function: 1 if input > 0, otherwise 0
def relu_deriv(x):
    return (x > 0).astype(float)  # Derivative of ReLU: returns 1 if x > 0, else 0

# Dataset: XOR problem
X = np.array([[0, 1], [1, 0], [1, 1], [0, 0]])  # Input data with 4 samples and 2 features
Y = np.array([[1], [1], [0], [0]])  # Expected output values for XOR function

# Initialize random weights for the layers
W1 = np.random.rand(2, 4)  # Weights from input layer (2 inputs) to first hidden layer (4 neurons)
W2 = np.random.rand(4, 3)  # Weights from first hidden layer (4 neurons) to second hidden layer (3 neurons)
W3 = np.random.rand(3, 1)  # Weights from second hidden layer (3 neurons) to output layer (1 neuron)

# Training loop (1000 steps)
for step in range(1000):  # Loop for 1000 iterations to train the model
    # Forward pass: Compute the activations for each layer
    Z1 = relu(np.dot(X, W1))  # First hidden layer: input X multiplied by W1, passed through ReLU
    Z2 = relu(np.dot(Z1, W2))  # Second hidden layer: Z1 multiplied by W2, passed through ReLU
    output = sigmoid(np.dot(Z2, W3))  # Output layer: Z2 multiplied by W3, passed through sigmoid activation

    # Calculate the error (difference between the predicted and actual outputs)
    error = Y - output  # Calculate the error (target Y minus predicted output)

    # Backward pass (Backpropagation)
    d_output = error * sigmoid_deriv(output)  # Gradient of the output layer (error * derivative of sigmoid)

    # Gradients for weights in output layer (W3)
    d_W3 = np.dot(Z2.T, d_output)  # Compute the gradient for W3 by multiplying Z2.T (transpose of Z2) with d_output

    # Gradients for the second hidden layer (Z2)
    d_Z2 = np.dot(d_output, W3.T) * relu_deriv(Z2)  # Compute gradient for Z2 (d_output * W3.T, passed through ReLU derivative)
    d_W2 = np.dot(Z1.T, d_Z2)  # Compute the gradient for W2 by multiplying Z1.T (transpose of Z1) with d_Z2

    # Gradients for the first hidden layer (Z1)
    d_Z1 = np.dot(d_Z2, W2.T) * relu_deriv(Z1)  # Compute gradient for Z1 (d_Z2 * W2.T, passed through ReLU derivative)
    d_W1 = np.dot(X.T, d_Z1)  # Compute the gradient for W1 by multiplying X.T (transpose of X) with d_Z1

    # Update weights using the gradients calculated during backpropagation
    W1 += d_W1  # Update weights for the first layer (input → first hidden)
    W2 += d_W2  # Update weights for the second layer (first hidden → second hidden)
    W3 += d_W3  # Update weights for the third layer (second hidden → output)

# After training, print the final output
print("Final Output:", output)  # Print the predicted output after training is completed
```

```python
# 21.Implement a simple Multi-Layer Perceptron with N binary inputs, two
# hidden layers and one output. Use backpropagation and Tanh function as
# activation function.
import numpy as np  # Importing the NumPy library for numerical operations

# Tanh activation function: returns tanh of the input (outputs values between -1 and 1)
def tanh(x):
    return np.tanh(x)  # Tanh function maps the input to a range between -1 and 1

# Derivative of the Tanh activation function
def tanh_deriv(x):
    return 1 - np.tanh(x) ** 2  # Derivative of Tanh: 1 - tanh(x)^2

# XOR dataset (4 samples, 2 binary inputs, 1 binary output)
X = np.array([[0, 1], [1, 0], [1, 1], [0, 0]])  # Input data for XOR operation
Y = np.array([[1], [1], [0], [0]])  # Expected output data (XOR results)

# Randomly initialize weights for each layer of the network
W1 = np.random.rand(2, 4)  # Weights for the first hidden layer (from input layer to hidden layer 1)
W2 = np.random.rand(4, 3)  # Weights for the second hidden layer (from hidden layer 1 to hidden layer 2)
W3 = np.random.rand(3, 1)  # Weights for the output layer (from hidden layer 2 to the output)

# Training loop (1000 iterations)
for step in range(1000):  # Run 1000 steps to train the model
    # Forward pass: Compute activations for each layer
    Z1 = tanh(np.dot(X, W1))  # First hidden layer: Input data (X) multiplied by weights (W1) and passed through Tanh activation
    Z2 = tanh(np.dot(Z1, W2))  # Second hidden layer: Z1 multiplied by weights (W2) and passed through Tanh activation
    output = sigmoid(np.dot(Z2, W3))  # Output layer: Z2 multiplied by weights (W3) and passed through sigmoid activation

    # Calculate error: Difference between the expected and actual outputs
    error = Y - output  # Error is the difference between the expected output (Y) and the predicted output

    # Backward pass (Backpropagation)
    d_output = error * sigmoid_deriv(output)  # Derivative of the output: error multiplied by derivative of the sigmoid activation

    # Gradient for weights at the output layer (W3)
    d_W3 = np.dot(Z2.T, d_output)  # Gradient for W3 is computed by multiplying the transpose of Z2 with the derivative of the output

    # Backpropagate the error to the second hidden layer (Z2)
    d_Z2 = np.dot(d_output, W3.T) * tanh_deriv(Z2)  # Compute the error propagated to Z2 (multiply by W3.T and apply tanh derivative)
    d_W2 = np.dot(Z1.T, d_Z2)  # Gradient for W2 is computed by multiplying the transpose of Z1 with d_Z2

    # Backpropagate the error to the first hidden layer (Z1)
    d_Z1 = np.dot(d_Z2, W2.T) * tanh_deriv(Z1)  # Compute the error propagated to Z1 (multiply by W2.T and apply tanh derivative)
    d_W1 = np.dot(X.T, d_Z1)  # Gradient for W1 is computed by multiplying the transpose of X with d_Z1

    # Update weights by adding the computed gradients (learning rate is omitted for simplicity)
    W1 += d_W1  # Update weights for the first layer (input to hidden layer 1)
    W2 += d_W2  # Update weights for the second layer (hidden layer 1 to hidden layer 2)
    W3 += d_W3  # Update weights for the output layer (hidden layer 2 to output)

# After training is complete, print the final output of the network
print("Final Output:", output)  # Display the network's output after 1000 training steps
```

```python
#23. Write a program to read a text file with at least 30 sentences and 200 words
# and perform the following tasks in the given sequence.
# Text cleaning by removing punctuation/special characters, numbers
# and extra white spaces. Use regular expression for the same.
# Convert text to lowercase
# Stemming and Lemmatization
# Create a list of 3 consecutive words after lemmatization
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Read the file
with open("text.txt", "r") as file:
    text = file.read()

# Text cleaning: Remove punctuation, special characters, numbers, and extra spaces using regex
text = re.sub(r'[^a-zA-Z\s]', '', text)  # Removes punctuation, numbers, and special characters
text = re.sub(r'\s+', ' ', text)  # Replaces multiple spaces with a single space
text = text.strip()  # Removes leading and trailing spaces
text = text.lower()  # Convert text to lowercase

# Tokenization
tokens = word_tokenize(text)

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Stemming and Lemmatization
stemmed = [stemmer.stem(word) for word in filtered_tokens]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# Create list of 3 consecutive words (bigrams) after lemmatization
three_word_sequences = [' '.join(lemmatized[i:i+3]) for i in range(len(lemmatized)-2)]

# Print outputs
print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
print("3 consecutive words after lemmatization:", three_word_sequences)
```
